Import packages

```python
import numpy as np
import skimage
import skimage.io
import matplotlib.pyplot as plt
```

```python
image = skimage.io.imread('https://cildata.crbs.ucsd.edu/media/images/13901/13901.tif')   #import image
```

```python
image
```

```python
image.shape
```

```python
plt.imshow(image, cmap='gray');   # 0- White , 1 - Black
```

# Accessing part of the array

```python
image[400:530,400:600]
```

```python
cropped = image[400:530,400:600].copy()
```

```python
plt.imshow(cropped, cmap='gray');
```

```python
cropped.shape
```

```python
image[400:530,:].shape
```

# Operation between arrays

```python
noise_image = np.random.poisson(lam=3 , size=cropped.shape)
```

```python
noise_image
```

```python
corrupted = cropped + 100*noise_image
```

```python
plt.imshow(cropped, cmap='gray')
```

```python
plt.imshow(corrupted, cmap='gray')
```

# Display images

```python
plt.imshow(np.random.randint(0,255,(10,10)))
```

```python
plt.imshow(np.random.choice([0,255],(10,10)))
```

```python
plt.imshow(np.random.choice([0,255],(10,10),p=[0.9,0.1]))    # 90% of 0 & 10% of 255
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.imshow(np.random.choice([0,128,255],(10,10),p=[0.4,0.5,0.1]))
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Random Image")
```
